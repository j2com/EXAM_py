# Datetime => Index로 설정 실습 ----------

- set_index(컬럼) : 인덱스

```python
# 모듈 로딩 ------------------------------
import pandas as pd

# 파일 관련 변수들 --------------------------------
DIR = '../Data/'
FILE_NAME = DIR+'stock-data.csv'
```

```python
stockDF = pd.read_csv(FILE_NAME)
```

```python
#(2) 데이터 확인
stockDF.head(3)
```

```python
stockDF.isnull().sum()
```

```python
# (3) 전처리 => NaN, Duplicate, DateType
# Date컬럼 Object => datetime64
stockDF['Date'] = stockDF['Date'].astype('datetime64')
```

```python
# 방법(2) pandas_to_datetime()
stockDF['Date']
```

```python
stockDF.head()
```

```python
# 특정 데이터 추출
stockDFdate = stockDF['Date']
stockDFdate[stockDFdate.dt.month>=7]
```

```python
stockDFdate[(stockDFdate.dt.month==6)&(stockDFdate.dt.day>=20)
            &(stockDFdate.dt.day<=25)]
```

```python
# index 변경 => Date 컬럼을 인덱스 설정
stockDF.set_index('Date', inplace=True)
stockDF
```

```python
stockDF.index
```

```python
stockDF.loc[stockDF.index[(stockDFdate.dt.month==6)&
            (stockDFdate.dt.day>=20)&
            (stockDFdate.dt.day<=25)],:]
```

```python
stockDF.loc['2018-06-20':'2018-06-25',:]
```

```python
stockDF.loc[stockDF.index[stockDFdate.dt.year==2018],:]
```
