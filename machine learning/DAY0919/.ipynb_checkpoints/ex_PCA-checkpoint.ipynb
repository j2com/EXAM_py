## PCA 주성분 분석
- iris의 4개 Feature 중에서 주성분 데이터 추출

```python
pca =PCA()
```

```python
# 지정된 데이터에서 주성분
pca.fit(X_scaled)
```

```python
pca.singular_values_
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

## 주성분 갯수 지정
- 방법1 :)explained_variance_ratio_ 의 누적값에서 원하는 분포정도 지정해서 PC갯수 계산
- 방법2 :)n_components = 0.0~1.0 : 주성분 투영된 데이터 분포 정도 지정하면 PC갯수 자동 계산

### 방법1 :)explained_variance_ratio_ 의 누적값으로 PC 차원수 계산

```python
import numpy as np
```

```python
cumsum = np.cumsum(pca.explained_variance_ratio_)
d=np.argmax(cumsum>=0.95)+1

print(f'주성분 갯수(차원): {d}')
```

```python
## 지정된 PC 갯수로 주성분 데이터 압축/융합
pca = PCA(n_components=d)
```

```python
pca.fit(X_scaled)
```

```python
irisPCA = pca.transform(X_scaled)
```
