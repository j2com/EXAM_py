```python
alpha = 'abcdefghijklmnopqrstuvwxyz'
```

```python
alpha_list = []
for i in alpha:
    alpha_list.append(i)
```

```python
alphaDF = pd.DataFrame(columns=alpha_list)
```

```python
alphaDF2 = pd.DataFrame(columns=alpha_list)
```

```python
def alpha_lang(filename):
    global alphaDF
    langname = pd.read_table('C:/Users/ss/Downloads/train-20220905T072451Z-001/train/'+filename, header=None)
    langname = langname.to_numpy()
    
    alpha_text = ""
    for i in langname:
        alpha_text = alpha_text + i[0]
    
    alpha_text = alpha_text.lower()
    
    alpha_count = []
    for i in alpha_list:
        alpha_count.append(alpha_text.count(i))
    for i in range(len(alpha_count)):
        alpha_count[i] = alpha_count[i]/sum(alpha_count) 
    alphaDF = alphaDF.append(pd.Series(alpha_count, index=alpha_list), ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.utils import all_estimators
```

```python
data = alphaDF.iloc[:,:-1]
target = alphaDF.iloc[:,-1]
```

```python
train_input, test_input, train_target, test_target = train_test_split(data, target, random_state= 42)
```

```python
models = all_estimators(type_filter='classifier')
```

```python
scores_train=[]
scores_test = []
model_name = []
for name, model in models:
    try:
        # 모델 객체 생성
        md = model()
        # 학습
        md.fit(train_input, train_target)
        # 평가
        scores_train.append(md.score(train_input, train_target))
        scores_test.append(md.score(test_input, test_target))
        model_name.append(name)
    except :
        pass
```
