# data, target 분류

```python
data = wineDF.iloc[:,:-1]
data
```

```python
target = wineDF.iloc[:,-1]
target
```

# train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

# all_estimators

```python
from sklearn.utils import all_estimators
```

```python
models = all_estimators(type_filter='classifier')
```

```python
scores = []
scores_train=[]
scores_test = []
model_name = []
for name, model in models:
    try:
        # 모델 객체 생성
        md = model()
        # 학습
        md.fit(train_input, train_target)
        # 평가
        scores.append(np.column_stack((name,md.score(train_input, train_target),md.score(test_input, test_target))))
        #scores_train.append(md.score(train_input, train_target))
        #scores_test.append(md.score(test_input, test_target))
        model_name.append(name)
    except :
        pass
```
