```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
fashion_mnist = tf.keras.datasets.fashion_mnist.load_data()
```

```python
train_input = fashion_mnist[0][0]
train_target = fashion_mnist[0][1]
test_input = fashion_mnist[1][0]
test_target = fashion_mnist[1][1]
```

```python
for i in range(0,10):
    plt.gray()
    plt.matshow(train_input[i])
    plt.show()
```

```python
train_input = train_input.reshape(-1,28*28)
test_input = test_input.reshape(-1,28*28)
```

```python
train_input.shape
```

```python
from sklearn.utils import all_estimators
```

```python
models = all_estimators(type_filter='classifier')
```

```python
scores_train=[]
scores_test = []
model_name = []
for name, model in models:
    try:
        # 모델 객체 생성
        md = model()
        # 학습
        md.fit(train_input, train_target)
        # 평가
        scores_train.append(md.score(train_input, train_target))
        #scores_test.append(md.score(test_input, test_target))
        #model_name.append(name)
    except :
        pass
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(train_input,train_target)
```

```python
print(lr.score(train_input, train_target))
print(lr.score(test_input, test_target))
```

# Ridge

```python
from sklearn.linear_model import Ridge
```

```python
Ridgemodel = Ridge(alpha=3.0)
```

```python
Ridgemodel.fit(train_input, train_target)
```

```python
print(Ridgemodel.score(train_input, train_target))
print(Ridgemodel.score(test_input, test_target))
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logis = LogisticRegression(max_iter=500)
```

```python
logis.fit(train_input, train_target)
```

```python
print(logis.score(train_input, train_target))
print(logis.score(test_input, test_target))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```
