```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
from matplotlib import font_manager as fm ,rc

DIR = '../../Data/'
df = pd.read_csv(DIR+'subwaytime1.csv', skiprows=1)
df.head()
```

```python
# 필요한 컬럼 추출
df1 = df.iloc[:,[1,3,11,13]]
df1
```

```python
# 하차 컬럼 int64 타입으로 변경
df1.iloc[:,2:4].astype('int64')
```

```python
# DF 컬럼명 바꾸기
df1 = df1.rename(columns={'Unnamed: 1':'호선명', 'Unnamed: 3': '지하철역', '하차.3':'07:00~07:59 하차', '하차.4':'08:00~08:59 하차'})
df1
```

```python
# 호선명 컬럼 셋 인덱스 후 7호선 까지 출력
subdf = df1.set_index('호선명')
subdf = subdf.loc[['1호선','2호선','3호선','4호선','5호선','6호선','7호선'],:]
subdf
```

```python
# 총 하차인원 컬럼 생성
subdf['하차인원'] = subdf['07:00~07:59 하차']+subdf['08:00~08:59 하차']
subdf
```

```python
# 호선명 reset index
subPartdf = subdf.reset_index('호선명')
```

```python
# 호선별 하차인원 기준 내림차수 정렬후 맨 처음 인덱스 출력 후 리스트 변환
subLinedf = 0
x_tick = []           # 그래프 x축
y_tick = []           # 그래프 y축
for i in range(1,8):
    subLinedf = subPartdf[subPartdf['호선명']==str(i)+'호선'].sort_values(by='하차인원', ascending=False)
    sub_list = subLinedf.iloc[0,:].to_list()
    x_tick.append(sub_list[0]+':'+sub_list[1])
    y_tick.append(sub_list[4])
    print(f'출근 시간대 {sub_list[0]} 최대 하차역은 {sub_list[1]}역이고 최대 하차인원은 {sub_list[4]:,}명 입니다.')

# 운영체제별 한글 폰트 설정
if(platform.system()=='Windows'):
    plt.rc('font',family='Malgun Gothic')
else:
    plt.rc('font',family='AppleGothic')
    
# 그래프 그리기
plt.figure(figsize=(20,8)) # 그래프 크기
plt.bar(x_tick,y_tick) # 막대그래프
plt.title('출근 시간대 지하철 노선별 최대 하차 인원 및 하차역') # 그래프 제목
plt.xticks(x_tick,rotation=80) # x축 레이블 지정하고 회전하기
plt.rc('font',size=20) # 폰트사이즈
plt.show() # 그래프 보여주기
```
