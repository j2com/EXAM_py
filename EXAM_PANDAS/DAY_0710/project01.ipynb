```python
list_2019 = []                                                  # 년도별 장르 데이터
list_2020 = []
list_2021 = []
list_2022 = []
for year, month in ymGroups.groups.keys():
    if year == 2019:
        list_2019.append(ymGroups.get_group((year, month)))
    elif year == 2020:
        list_2020.append(ymGroups.get_group((year, month)))
    elif year == 2021:
        list_2021.append(ymGroups.get_group((year, month)))
    elif year == 2022:
        list_2022.append(ymGroups.get_group((year, month)))
```
