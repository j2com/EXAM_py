```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from matplotlib import font_manager as fm, rc

font_path='../../Font/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
DIR = '../../Data/'
FILE1 = DIR+'애니(남자).xlsx'
FILE2 = DIR+'애니(여자).xlsx'
FILE3 = DIR+'영화(남자).xlsx'
FILE4 = DIR+'영화(여자).xlsx'
FILE5 = DIR+'드라마(남자).xlsx'
FILE6 = DIR+'드라마(여자).xlsx'
FILE7 = DIR+'예능(남자).xlsx'
FILE8 = DIR+'예능(여자).xlsx'
```

```python
FILEAGE1 = DIR+'18세이하.xlsx'
FILEAGE2 = DIR+'20대.xlsx'
FILEAGE3 = DIR+'30대.xlsx'
FILEAGE4 = DIR+'40대.xlsx'
FILEAGE5 = DIR+'50대.xlsx'
FILEAGE6 = DIR+'60세이상.xlsx'
```

```python
df1 = pd.read_excel(FILE1, skiprows=6)
df2 = pd.read_excel(FILE2, skiprows=6)
df3 = pd.read_excel(FILE3, skiprows=6)
df4 = pd.read_excel(FILE4, skiprows=6)
df5 = pd.read_excel(FILE5, skiprows=6)
df6 = pd.read_excel(FILE6, skiprows=6)
df7 = pd.read_excel(FILE7, skiprows=6)
df8 = pd.read_excel(FILE8, skiprows=6)
```

```python
ageDF1 = pd.read_excel(FILEAGE1, skiprows=6)
ageDF2 = pd.read_excel(FILEAGE2, skiprows=6)
ageDF3 = pd.read_excel(FILEAGE3, skiprows=6)
ageDF4 = pd.read_excel(FILEAGE4, skiprows=6)
ageDF5 = pd.read_excel(FILEAGE5, skiprows=6)
ageDF6 = pd.read_excel(FILEAGE6, skiprows=6)
```

```python
df1.rename(columns={'디즈니플러스':'Ani_Cnt(M)'}, inplace=True)
df2.rename(columns={'디즈니플러스':'Ani_Cnt(F)'}, inplace=True)
df3.rename(columns={'디즈니플러스':'Movie_Cnt(M)'}, inplace=True)
df4.rename(columns={'디즈니플러스':'Movie_Cnt(F)'}, inplace=True)
df5.rename(columns={'디즈니플러스':'Drama_Cnt(M)'}, inplace=True)
df6.rename(columns={'디즈니플러스':'Drama_Cnt(F)'}, inplace=True)
df7.rename(columns={'디즈니플러스':'Enter_Cnt(M)'}, inplace=True)
df8.rename(columns={'디즈니플러스':'Enter_Cnt(F)'}, inplace=True)
```

```python
allDF = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8], axis=1, join='inner') # concat으로 연결
```

```python
allDF = allDF.loc[:,~allDF.T.duplicated()] # 중복 컬럼 제거
```

```python
allDF['날짜'] = pd.to_datetime(allDF['날짜'])
```

```python
allDF['Year'] = allDF['날짜'].dt.year
allDF['Month'] = allDF['날짜'].dt.month
allDF['Day'] = allDF['날짜'].dt.day
allDF
```

```python
allDFsi = allDF.set_index('날짜') # 날짜를 index set index
allDFsi
```

```python
allDFsi.iloc[:,:2].plot(figsize=(15,6)) # 애니 남녀
```

```python
df2019 = allDF[allDF['날짜'].dt.year == 2019].set_index('날짜')
df2020 = allDF[allDF['날짜'].dt.year == 2020].set_index('날짜')
df2021 = allDF[allDF['날짜'].dt.year == 2021].set_index('날짜')
df2022 = allDF[allDF['날짜'].dt.year == 2022].set_index('날짜')    
```

```python
df2019_1 = df2019.reset_index().iloc[:,1:-1]     
df2020_1 = df2020.reset_index().iloc[:,1:-1]
df2021_1 = df2021.reset_index().iloc[:,1:-1]
df2022_1 = df2022.reset_index().iloc[:,1:-1]
```

```python
df2019group = df2019_1.groupby(['Year','Month']).mean()    # 년도, 월별 그룹화 후 평균
df2020group = df2020_1.groupby(['Year','Month']).mean()
df2021group = df2022_1.groupby(['Year','Month']).mean()
df2022group = df2022_1.groupby(['Year','Month']).mean()
```
