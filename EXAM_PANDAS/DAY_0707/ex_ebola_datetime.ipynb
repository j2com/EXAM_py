```python
import pandas as pd
from datetime import datetime
current = datetime.today()
DIR = '../../Data/'
FILENAME = DIR+'country_timeseries.csv'
```

```python
ebolaDF = pd.read_csv(FILENAME)
ebolaDF.head()
```

```python
#(2) 데이터 확인
ebolaDF.info()
```

```python
# (3) 전처리
ebolaDF = ebolaDF.dropna(how = 'all').fillna(method='ffill')
ebolaDF
```

```python
ebolaDF['Date'] = pd.to_datetime(ebolaDF['Date'])
ebolaDF['Date'].dtype
```

```python
d_days= current - ebolaDF['Date']
```

```python
d_days[0].days, d_days[0].delta, d_days[0].seconds
```

```python
ebolaDF['D-Day']= d_days
```
