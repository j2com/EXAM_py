#### 데이터 정규화

- 수치데이터 컬럼마다 값의 범위가 다름
- 값의 범위 0~1 / -1~1 로 값을 설정

##### (3-3-1) 방법 : 컬럼의 최대값의 절대값으로 전체 데이터를 나누기

```python
abs(autoDF['weight'].max()),autoDF['weight'].max(),autoDF['weight'].min()
```

```python
autoDF['weight_nor']=autoDF['weight'] / abs(autoDF['weight'].max())
autoDF['weight_nor'].min(), autoDF['weight_nor'].max()
```

```python
numerator = autoDF['weight'] - autoDF['weight'].max()
denominator = autoDF['weight'].max() - autoDF['weight'].min()

autoDF['weight_nor2'] = numerator/denominator
autoDF['weight_nor2']
```

##### (3-3-3) 방법 : 컬럼의 데이터 - 평균 / 표준편차 ==> Z.score

```python
numerator = autoDF['weight'] - autoDF['weight'].mean()
denominator = autoDF['weight'].std()

autoDF['weight_nor_3']=numerator/denominator
autoDF['weight_nor_3'].min(), autoDF['weight_nor_3'].max()
```

```python
autoDF[['weight', 'weight_nor2', 'weight_nor_3']]
```

```python
def calCol(colName):
    
    for i in range(0,len(autoDF.columns)):
        if autoDF.columns[i] == colName:
            numerator = autoDF[colName] - autoDF[colName].max()
            denominator = autoDF[colName].max() - autoDF[colName].min()

            autoDF[colName+'nor'] = numerator/denominator
    return autoDF[colName+'nor']
calCol('displacement')
autoDF
        
```
