```python
import csv
import matplotlib.pyplot as plt
import platform
import matplotlib.font_manager as fm
import math
import re

DIR = '../../Data/'
f = open(DIR+'gender.csv', encoding='euc_kr')
data = csv.reader(f)
```

```python
def parse_city_name(city_1):
    '''
    행정구역명에서 도시 이름 파싱 (코드 번호 제거)
    :param city:
    :return:
    '''
    city_name = re.split('[()]', city_1)
    # [0]: 행정구역 이름, [1]:
    return city_name[0]
```

```python
def cityPart(district):                                # 지역 이름 받기
    f = open(DIR+'gender.csv', encoding='euc_kr')
    data = csv.reader(f)
    city_name = ''
    male_num_list = []
    female_num_list = []
    for city in data:
        if '대구광역시' in city[0]:                            # 대구광역시 + distict 있으면 실행
            if district in city[0]:
                city_name = parse_city_name(district)         # city_name에 행정구역 이름 
                for male in city[104:207]:
                    if ',' in male:
                        male = male.replace(',','')           
                    male_num_list.append(int(male))           # 남자 수 리스트에 담기
                for female in city[207:]:
                    if ',' in female:
                        female = female.replace(',','')
                    female_num_list.append(int(female))       # 여자 수 리스트에 담기
    sum_male = sum(male_num_list)                             # 남자 총 인원
    sum_female = sum(female_num_list)                         # 여자 총 인원
    f.close()

    draw_piechart(city_name, sum_male, sum_female)            # 파이차트 그리기
```

```python
def draw_piechart(city_name, sum_male, sum_female): # 파이 차트 함수
    '''
    전체 인구수 대비 투표 가능 인구의 파이차트 작성
    '''
    population = [sum_male, sum_female]
    if platform.system() == 'Windows':
        plt.rc('font', family='Malgun Gothic')
    else:
        plt.rc('font', family='AppleGothic')
    color = ['tomato', 'royalblue']
    plt.pie(population, labels=['남성', '여성'], autopct='%.1f%%',
            colors=color, startangle=90)

    plt.title(city_name)
```

```python
# 전체 subplot의 개수 설정 (4 x 2= 총 8개)
plt.subplots(2, 4, figsize=(15, 8), sharex=True, sharey=True)
# 전체 그래프의 타이틀 설정                      share = x,y 축 공유
plt.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=15)
plt.subplot(2,4,1)
cityPart('대구광역시 중구')

plt.subplot(2,4,2)
cityPart('대구광역시 동구')

plt.subplot(2,4,3)
cityPart('대구광역시 서구')

plt.subplot(2,4,4)
cityPart('대구광역시 남구')

plt.subplot(2,4,5)
cityPart('대구광역시 북구')

plt.subplot(2,4,6)
cityPart('대구광역시 수성구')

plt.subplot(2,4,7)
cityPart('대구광역시 달서구')

plt.subplot(2,4,8)
cityPart('대구광역시')

plt.show()
```
