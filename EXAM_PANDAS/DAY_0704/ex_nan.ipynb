# 결측치(Missing Value) 처리 실습-------

- 결측치(Missing Value)란?
    : 값이 들어있지 않는 것으로 NaN, Na로 표기
    
- 관련 메서드
    - 체크 : isnull(), notnull(), isna(), notna()
    - 삭제 : dropna()
    - 치환 : fillna()

```python
import pandas as pd
DIR_PATH = '../../Data/'
FILR_NAME = DIR_PATH + 'weather.csv'

dfWeather = pd.read_csv(FILR_NAME)
dfWeather
```

```python
dfWeather.info()
```

```python
# 결측치 갯수 파악 ------------------------------------
dfWeather.isnull().sum() #True 값 더해서 sum
```

```python
oneMonth = dfWeather.iloc[:2,]
oneMonth
```

```python
oneMonth.dropna(axis=1, how = 'all')
```

```python
# 2월의 최고 - 최저온도 출력
twoMonth = dfWeather.iloc[2:4,].dropna(axis=1, how= 'all')
twoMonth
```
