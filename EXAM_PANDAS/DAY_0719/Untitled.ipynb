```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

DIR = '../../Data/'
df = pd.read_excel(DIR+'월간 액화천연가스 수입 .xlsx', skiprows=7)
priceDF = pd.read_csv(DIR + '천연가스 선물 내역.csv')
eurpriceDF = pd.read_csv(DIR+'유럽 천연가스.csv')
```

```python
df.columns
```

```python
df2 = df.loc[:,'2020':]
```

```python
df1 = df.iloc[:,1]
```

```python
naDF = pd.concat([df1,df2], axis=1)
naDF = naDF.iloc[2:,:]
naDF
```

```python
naDF.rename(columns={'항목1*.1':'항목'}, inplace=True)
```

```python
naDF
```

```python
naDF = naDF.set_index('항목')
```

```python
new_col = pd.date_range(start='2020-1', end='2022-5', freq='M').strftime('%Y.%m').to_list() # 컬럼명 변경
naDF.columns = new_col
naDF = naDF.loc[:,'2021.04':]
naDF
```

```python
naDF_T = naDF.T
naDF_T['러시아 비율'] = naDF_T['합계(1,000 톤)'] / naDF_T['러시아(1,000 톤)']
naDF_T
```

```python
naDF_T['count'] = 100
naDF_T['러시아 제외 국가 비율'] = naDF_T['count'] - naDF_T['러시아 비율']
naDF_T
```

```python
naDF2 = naDF.iloc[2:,:]
plt.rc('font',family='Malgun Gothic')
naDF2.T.plot(figsize = (20,8))
plt.title('한국 천연가스 수입량')
```

```python
naDF_T.loc[:,['러시아 제외 국가 비율','러시아 비율']].plot.bar(stacked=True, figsize=(20,7))
```

```python
naDF.iloc[:2,:].T.plot(figsize = (20,8))
plt.title('한국 천연가스 총 수입량')
```
